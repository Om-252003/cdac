```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
# Document Term Matrix
```

```python
d1 = "First Document"
d2 = "Second Document which is not like the first document means it is different from it"
d3 = "Third Document which unlike like both first and second"
```

```python
content=[d1,d2,d3]
cv=CountVectorizer()
m=cv.fit_transform(content)
```

```python
m.toarray()
```

```python
import pandas as pd
df=pd.DataFrame(m.toarray(),columns=cv.get_feature_names_out())
df
```

```python
content=[d1,d2,d3]
my_stop_words=['and','is','i','the','that','this','from','my']

cv=CountVectorizer(stop_words=my_stop_words)
m=cv.fit_transform(content)
```

```python
all_content = " ".join(content)
cv_all = CountVectorizer()
m_all = cv_all.fit_transform([all_content])
df_all = pd.DataFrame(m_all.toarray(), columns=cv_all.get_feature_names_out())
word_counts = df_all.sum().sort_values(ascending=False)
print(word_counts)
```

```python
content = [d1,d2,d3]
tfidf = TfidfVectorizer(stop_words='english')
m = tfidf.fit_transform(content)
df = pd.DataFrame(m.toarray(), columns=tfidf.get_feature_names_out())
print(df)
```

```python
df = pd.read_csv('/content/SMSSpamCollection', sep="\t", names=['Status','Message'])
df.info()
```

```python
df['Status'].value_counts()
```

```python
# mask = df['Status'] == 'Spam'
# df.loc[mask, 'Status'] = 1

# mask = df['Status'] == 'ham'
# df.loc[mask, 'Status'] = 0

# df.head()
```

```python
x = df['Message']
y = df[['Status']]
```

```python
# preprocessing of x
cv1 = CountVectorizer()  # arguments can be lowercase=boolean, stopwords='english', etc.
xnew = cv1.fit_transform(x)
xnew.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(xnew,y,test_size=0.2,random_state=1)
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
# accepts output from CounVectorizer and TfidfVectorizer
# but efficient on int than float

mnb = MultinomialNB()
mnb.fit(xtrain,ytrain)
```

```python
yhat = mnb.predict(xtest)
print(classification_report(ytest,yhat))
```
