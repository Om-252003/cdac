# Lasso and Ridge Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("/content/50_Startups.csv")
df.info()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
X_numerical = df[['RDSpend', 'Administration', 'MarketingSpend']]
X_state = pd.get_dummies(df['State'], drop_first=True, dtype=int)

x = pd.concat([X_numerical, X_state], axis=1).values
y = df.iloc[:,[-1]].values

print(x.shape,y.shape)
```

```python
from sklearn.linear_model import LassoCV, RidgeCV, ElasticNetCV
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=90)
```

```python
model = LassoCV()
model.fit(xtrain,ytrain).score(xtest,ytest)
# .score internally calls predict and r2_score methods
```

```python
model.coef_
```

```python
model2 = RidgeCV()
model2.fit(xtrain,ytrain)
print(model2.score(xtest,ytest))
np.set_printoptions(suppress=True)
model2.coef_
```

```python
model3 = ElasticNetCV()
model3.fit(xtrain,ytrain).score(xtest,ytest)
```

```python
model3.coef_
```

ElasticNet : Linear Regression + Lasso + Ridge as regularizer

CV menas crossfold validation

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=100)

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
model = LinearRegression()
accuracy = cross_val_score(model,xtrain,ytrain,cv=5)
print(accuracy)
print("Average training Accuracy ",np.mean(accuracy).round(2),
      ' with SD of : ', np.std(accuracy).round(2) )
```

```python
from sklearn.metrics import r2_score
model.fit(xtrain,ytrain)
round(r2_score(ytest,model.predict(xtest)),2)
```
