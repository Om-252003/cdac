```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
sc = spark.sparkContext
```

```python
list1 = [10,20,30,40,50]
rdd1 = sc.parallelize(list1)
```

```python
print("Printing all elements of the rdd",rdd1.collect())

print("Printing the sum of all elements of the rdd",rdd1.sum())

print("Printing the max values among all elements of the rdd",rdd1.max())

print("Printing the min values among all elements of the rdd",rdd1.min())

print("Printing the min values among all elements of the rdd",rdd1.mean())

print("Printing the count of elements of the rdd",rdd1.count())

print("Printing the first element of the rdd",rdd1.first())

print("Printing the first 3 element of the rdd",rdd1.take(3))

print("Printing the standard deviation of elements of the rdd",rdd1.stdev())

print("Printing the variance of elements of the rdd",rdd1.variance())

print("Printing the stats of elements of the rdd",rdd1.stats())
```
