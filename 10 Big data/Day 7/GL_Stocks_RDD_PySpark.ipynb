```pyspark
import pyspark
sc = spark.sparkContext
```

```pyspark

filePath = '/user/hadoop/NYSE_Data.txt'

stocksRDD = sc.textFile(filePath)

maxClosePriceRDD = stocksRDD.map(lambda x: (x.split(',')[1], x.split(',')[6]))
```

```pyspark
myRDD = maxClosePriceRDD.reduceByKey(lambda a,b: max(a,b))
```

```pyspark
myRDD.collect()
```
