```python
class Student:
    def __init__(self,name,age):
        self.name=name
        self.age=age
           
    def getname(self):
        return self.name
        
    def getage(self):
        return self.age

    def __str__(self):
        return ""+ self.name+"  and  "+str(self.age)
```

```python
s1 = Student('Vishal',23)
s2 = Student('Sachin',32)
s3 = Student('Anands',26)

print(s1)

students = {s1:100,s2:200,s3:50}
```

```python

for i in students:
    print(i)
```

```python

mylist= sorted(students,key=Student.getname)

print(type(mylist))
print("Namewise sort")
for i in mylist:
    print(i)
```

```python

mylist = sorted(students,key=Student.getage)
print(type(mylist))
print("Agewise sort")
for i in mylist:
    print(i)

```

```python

There are 3 ways to merge dictionaries

The update() method—dict1.update(dict2)—updates an existing
dictionary.

{**dict1, **dict2} creates a new merged dictionary using unpacking.
{1:'Anil',2:'Rohit',3:'Vishal' , 'Maths':300,"science":400,"History":200 }

The union | operator in Python 3.9+ merge dictionaries 
 simple syntax dict1 | dict2.
```

```python

mydict1={1:'Anil',2:'Rohit',3:'Vishal'}
mydict2={'Maths':300,"science":400,"History":200}

print("mydict1 is\t",mydict1)
print("mydict2 is\t",mydict2)
```

```python


mydict1.update(mydict2)
print(mydict1)

```

```python


mydict3=mydict1 | mydict2
print(mydict3)
```

```python
mydict3={**mydict1,**mydict2}
print(mydict3)
```
