```python
from sklearn.feature_extraction.text import CountVectorizer
```

Document Term Matrix

```python
d1="This is my first document"
d2="This is my second document"
d3="This is my third document"
```

```python
content=[d1,d2,d3]
cv=CountVectorizer()
m=cv.fit_transform(content)
```

```python
m.toarray()
```

```python
df=pd.DataFrame(m.toarray(),columns=cv.get_feature_names_out())
```

```python
df
```

```python
content=[d1,d2,d3]
my_stop_words=['and','is','i','the','that','this','from','my']
cv=CountVectorizer(stop_words=my_stop_words)
m=cv.fit_transform(content)
```

```python
df=pd.DataFrame(m.toarray(),columns=cv.get_feature_names_out())
```

```python
df
```

```python
all_content = " ".join(content)
cv_all = CountVectorizer()
m_all = cv_all.fit_transform([all_content])
df_all = pd.DataFrame(m_all.toarray(), columns=cv_all.get_feature_names_out())
word_counts = df_all.sum().sort_values(ascending=False)
print(word_counts)
```
