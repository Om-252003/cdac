```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/SLRAssignment1_Dataset.xlsx")
df
```

```python
# Question :  Find the coefficient and the intercept and test it with some values. Target is y. Predictor is X
```

```python
df.info()
```

```python
df[['X','y']].corr()['y']
```

```python
plt.figure(figsize=(5,3))
plt.scatter(df['X'],df['y'])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[['X']]
y = df[['y']]
print(X.shape, y.shape)
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
test_vals = np.array([[100],[60]]).reshape(-1,1)
test_vals.shape
```

```python
yhat = model.predict(test_vals)
print(np.round(yhat, 2))
```

```python
plt.figure(figsize=(5,3))
plt.title('Linear Regression')
plt.scatter(df['X'], df['y'], c='red', label='Actual', marker='d', s=100)
plt.scatter(df['X'], model.predict(X), c='blue', label='Predicted')
plt.plot(df['X'], model.predict(X), c='green', label='Regression Line')
plt.legend(bbox_to_anchor=(1.2,1))
plt.show()
```
